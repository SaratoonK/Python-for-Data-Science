## 1. Let's create a sine wave

```python
#let t = 1 to 800
#f = 1/40 frequency
t = torch.linspace(0, 799, steps = 800)
y = torch.sin(2 * torch.pi * t / 40) #2*pi*t / f

plt.figure(figsize=(12,4))
plt.plot(y)

```
